```python
def select_position():
    column_choice = ' '
    row_choice = ' '
    acceptable = ['0','1','2']
    
    #Player will choose their mark's position on the board
    #this loops till we get the acceptable choice from the user
    while (row_choice not in acceptable) or (column_choice not in acceptable):
        row_choice = input('Enter row (0,1,2): ')
        column_choice = input('Enter column (0,1,2): ')
        
        #wrong choice then throw error message
        if (row_choice not in acceptable) or (column_choice not in acceptable):
            print('You did not enter corect row ot column. Please try again!')
            
    return int(row_choice), int(column_choice)
```

```python
def update_board():
    row_choice,column_choice = select_position()
    
    #updating the board with the player's mark at player's board position choice 
    if select_player == 'player 1':
        if grid[row_choice][column_choice] == '':
            grid[row_choice][column_choice] = player_1_mark
        else:
            print('SELECTED POSITION ALREADY FILLED!')
    
    elif select_player == 'player 2':
        if grid[row_choice][column_choice] == '':
            grid[row_choice][column_choice] = player_2_mark
        else:
            print('SELECTED POSITION ALREADY FILLED!')
    
    return grid
```

```python
def display_board():
    #clears the output after every update of board before displaying the updated board
    clear_output()
    
    #displaying the board
    print(grid[0][0],'|',grid[0][1],'|',grid[0][2])
    print('--','--','--')
    print(grid[1][0],'|',grid[1][1],'|',grid[1][2])
    print('--','--','--')
    print(grid[2][0],'|',grid[2][1],'|',grid[2][2])
```

```python
def winner_check(mark):
    
    if ((grid[0][0] == grid[0][1] == grid[0][2] == mark) or #row 1 check 
        (grid[1][0] == grid[1][1] == grid[1][2] == mark) or #row 2 check
        (grid[2][0] == grid[2][1] == grid[2][2] == mark) or #row 3 check
        (grid[0][0] == grid[1][0] == grid[2][0] == mark) or #column 1 check
        (grid[0][1] == grid[1][1] == grid[2][1] == mark) or #column 2 check
        (grid[0][2] == grid[1][2] == grid[2][2] == mark) or #column 3 check
        (grid[0][0] == grid[1][1] == grid[2][2] == mark) or #diagonal 1 check
        (grid[2][0] == grid[1][1] == grid[0][2] == mark) #diagonal 2 check
       ):
    
        return True
    
    else:
        return False
```

```python
def board_check():
    
    #checks if the board has any empty places, if any empty place then sends false to continue the game
    if ((grid[0][0] == '' or grid[0][1] == '' or grid[0][2] == '') or 
       (grid[1][0] == '' or grid[1][1] == '' or grid[1][2] == '') or
        (grid[2][0] == '' or grid[2][1] == '' or grid[2][2] == '')
       ):
        return False
    else:
        print('Game draw!')
        return True
```

```python
from IPython.display import clear_output

#board
grid = [['','',''],['','',''],['','','']]

#select player's choice of symbol
mark = ['X','O']
select_mark = '$'

while (select_mark not in mark):
    select_mark = input('Player 1, select your mark (X or O):' ).upper()

    if select_mark not in mark:
        print('Please select valid mark X or O')

#assign player 1's and player 2's marks

player_1_mark = select_mark

if player_1_mark == 'X':
    player_2_mark = 'O'
else:
    player_2_mark = 'X'
        
        
print('_____________________________________') 
print(' ')
print('player 1: ',player_1_mark,'|','player 2: ', player_2_mark)
print('_____________________________________')     


move = 1
while 1:
    #here if move is odd consider player 1 is playing, if even - player 2 is playing
    if move%2 == 0:
        select_player = 'player 2'
    else:
        select_player = 'player 1'
    
    print(' ')
    print(select_player,': ')
    
    #To choose postion on board (this function will be called inside update_board func) and update the board
    update_board()
    
    #display the board
    display_board()
    
    #increment move
    move +=1
    
    
    #if winner or board filled, end game
    
    if winner_check(player_1_mark):
        print('Player 1 Wins!!')
        print('Game over!')
        break
    
    elif winner_check(player_2_mark):
        print('Player 2 Wins!!')
        print('Game over!')
        break
        
    elif board_check():
        break
```
